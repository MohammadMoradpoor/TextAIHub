```python
for i in tqdm(range(1, 3231)):
    page = requests.get('https://ganjoor.net/moulavi/shams/ghazalsh/sh' + str(i) + '/')
    tree = html.fromstring(page.content)
    with open('moulavi.txt', 'a', encoding="utf-8") as f:
        for t in tree.xpath('//div[@class="b"]/div/p/text()'):
            f.write('|' + str(t.replace('\u200c', ' ')) + '\n')
```
